```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_theme()
```

```python
#Se importa la hoja HECHOS_FALLECIDOS del archivo de EXCEL de homicidios 

nombre_archivo = 'homicidios.xlsx'
nombre_hoja = 'HECHOS_FALLECIDOS'
df=pd.read_excel(nombre_archivo,sheet_name=nombre_hoja)
df.head(10)
```

```python
df.info()
```

```python
#Se modifica el tipo de dato de las columnas categóricas

df['LUGAR_DEL_HECHO']      = df['LUGAR_DEL_HECHO'].astype('category')
df['TIPO_DE_CALLE']        = df['TIPO_DE_CALLE'].astype('category')
df['Calle']                = df['Calle'].astype('category')
df['Altura']               = df['Altura'].astype('category')
df['Cruce']                = df['Cruce'].astype('category')
df['Dirección Normalizada']= df['Dirección Normalizada'].astype('category')
df['COMUNA']               = df['COMUNA'].astype('category')
df['PARTICIPANTES']        = df['PARTICIPANTES'].astype('category')
df['VICTIMA']              = df['VICTIMA'].astype('category')
df['ACUSADO']              = df['ACUSADO'].astype('category')

df.info()

```
